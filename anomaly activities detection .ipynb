# Network Intrusion Detection

#### Importing Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score,recall_score,accuracy_score
```

```python
# importing the data into pandas DataFrame
training_data = pd.read_csv('Train_Data.csv')
training_data.head() # Showing the first 5 rows
```

```python
# Lets see the Description of the data
training_data.describe()
```

```python
# Lets check the statistical information of the data
training_data.info
```

```python
# Lets check if the data have any mising values
training_data.isnull().sum()
```

That's great, our dataset don't have any missing values

From the statistical information, we got to know that we have some categorical data
- these categorical features needs to be converted into numerical values
- because machine learning model don't understand categorical data
- we will use the getdummies() method to convert the data


```python
training_data = pd.get_dummies(training_data,columns = ["protocol_type", "flag", "service"], drop_first=True)

training_data.head()
```

we have converted protocol_type, flag, service into numerical values

```python
# Lets check the class feature
training_data['class'].unique()
```

The class feature is also categoricategorical, we will convert it into 0 and 1
- 0 for Normal
- 1 for anomaly

```python
# creating a dict file 
Class = {'normal': 0,'anomaly': 1}
# traversing through dataframe
training_data['class'] = [Class[item] for item in training_data['class']]
training_data.head()
```

```python
training_data['class'].unique()
```

we can see that we have successfully converted the string values into numerical values

```python
# lets check how many rows are normal and anomalies 
training_data['class'].value_counts()
```

#### splitting the data

- we are done with basic data analysis needed
- we will now split the data into testing and training sets
- the training data will be used to train the models
- testing data will be used to test the models 

```python
X = training_data.drop('class', axis = 1)
Y = training_data['class']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2,random_state=0)
```

#### Model Building

we have splitted the data into test and train sets, we will now train our classification models 
- we will train the following models
- Logistic Regression
- Decision Tree Classifiers
- Random Forests
- Support Vector Machines
- K-Nearest Neighbors
- Gaussian Naive Bayes
- LinearDiscriminantAnalysis

```python
df_perf_metrics = pd.DataFrame(columns=[
    'Model', 'Accuracy_Training_Set', 'Accuracy_Test_Set', 'Precision',
    'Recall', 'f1_score'
])
models_trained_list = []


def get_perf_metrics(model, i):
    # model name
    model_name = type(model).__name__
    print("Training {} model...".format(model_name))
    # Fitting of model
    model.fit(X_train, y_train)
    print("Completed {} model training.".format(model_name))
    # Predictions
    y_pred = model.predict(X_test)
    # Add to ith row of dataframe - metrics

    df_perf_metrics.loc[i] = [
        model_name,
        model.score(X_train, y_train),
        model.score(X_test, y_test),
        precision_score(y_test, y_pred),
        recall_score(y_test, y_pred),
        f1_score(y_test, y_pred),
    ]
   
    print("Completed {} model's performance assessment.".format(model_name))
```

```python
# the model list
models_list = [LogisticRegression(),
               DecisionTreeClassifier(),
               RandomForestClassifier(),
               SVC(),
               KNeighborsClassifier(),
               GaussianNB(),LinearDiscriminantAnalysis()
               ]
```

```python
from sklearn.metrics import r2_score,f1_score
for n, model in enumerate(models_list):
    get_perf_metrics(model, n)
```

```python
df_perf_metrics
```

Above is the accuracy of different models that we have trained 
- lets take Random Forest and perform parameter tuning

#### Random Forest Parameter tuning

```python
rfc = RandomForestClassifier()
parameters = {
    "n_estimators":[5,10,50,100,250],
    "max_depth":[2,4,8,16,32,None]}
```

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(rfc,parameters,cv=5)
cv.fit(X_train,y_train.values.ravel())
```

```python
def display(results):
    print(f'Best parameters are: {results.best_params_}')
    print("\n")
    mean_score = results.cv_results_['mean_test_score']
    std_score = results.cv_results_['std_test_score']
    params = results.cv_results_['params']
    for mean,std,params in zip(mean_score,std_score,params):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
```

```python
display(cv)
```

```python
model = cv.best_estimator_
y_pred = model.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Precision: ', precision_score(y_test, y_pred))
print('Recall: ', recall_score(y_test, y_pred))
print('f1-score: ', f1_score(y_test, y_pred))
```
